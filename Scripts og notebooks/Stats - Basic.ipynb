```python
import numpy as np

Float = [444,479,480]
Grid =[593,465,529,535,532,540,560,514,484]
OneHot=[404,466,449,421,401,405,428,474,448,464]

print(f"{np.mean(Float):.3f} , {np.std(Float):.3f} , {np.mean(Grid)} , {np.std(Grid):.3f} , {np.mean(OneHot)} , {np.std(OneHot):.3f}")

# Confidence interval for Float
n = len(Float)
mean = np.mean(Float)
std = np.std(Float)
t =  2.920 # t-value for 95% confidence interval and n=3
CI = t * std / np.sqrt(n)
print(f"Float: [{mean-CI/2:.3f} : {mean+CI/2:.3f}]")

# Confidence interval for Grid
n = len(Grid)
mean = np.mean(Grid)
std = np.std(Grid)
t = 1.860 # t-value for 95% confidence interval and n=9
CI = t * std / np.sqrt(n)
print(f"Grid: [{mean-CI/2:.3f} : {mean+CI/2:.3f}]")

# Confidence interval for OneHot
n = len(OneHot)
mean = np.mean(OneHot)
std = np.std(OneHot)
t = 1.833 # t-value for 95% confidence interval and n=10
CI = t * std / np.sqrt(n)
print(f"OneHot: [{mean-CI/2:.3f} : {mean+CI/2:.3f}]")
```
